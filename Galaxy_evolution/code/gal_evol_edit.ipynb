```python
color1_before1 = cosmos2020_bright_mask4['HSC_g_MAG_AUTO'] - cosmos2020_bright_mask4['HSC_i_MAG_AUTO'] #g-i
color2_before2 = cosmos2020_bright_mask4['HSC_r_MAG_AUTO'] - cosmos2020_bright_mask4['HSC_z_MAG_AUTO'] #r-z
color3_before3 = cosmos2020_bright_mask4['CFHT_u_MAG_AUTO'] - cosmos2020_bright_mask4['HSC_r_MAG_AUTO'] #u-r
color_before = np.vstack([color1_before1, color2_before2, color3_before3]).T
```

```python
color1_after1 = cosmos2020_bright_mask5['HSC_g_MAG_AUTO'] - cosmos2020_bright_mask5['HSC_i_MAG_AUTO'] 
color2_after2 = cosmos2020_bright_mask5['HSC_r_MAG_AUTO'] - cosmos2020_bright_mask5['HSC_z_MAG_AUTO'] #r-z
color3_after3 = cosmos2020_bright_mask5['CFHT_u_MAG_AUTO'] - cosmos2020_bright_mask5['HSC_r_MAG_AUTO'] #u-r
color_after = np.vstack([color1_after1, color2_after2, color3_after3]).T
```

```python
labels = ['g-i', 'r-z', 'u-r']
c1 = corner.corner(color_before, color="blue")
corner.corner(color_after,color = "red")

plt.show() 
```

```python
c = corner.corner(color_before, labels=labels, color = "blue", label_kwargs={"fontsize":12}, title_fmt=".2f", show_titles=True, 
                   hist_kwargs= {"density":True}   )
corner.corner(color_after, labels=labels, fig = c, color = "red", title_fmt=".2f", show_titles=True, hist_kwargs={"density":True})
plt.show()
```
